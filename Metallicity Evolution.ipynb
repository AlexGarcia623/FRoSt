```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LogNorm

mpl.rcParams['font.size'] = 15
```

$$\frac{dZ_*}{dt} = \frac{1}{M_*}\frac{dM_*}{dt} \left(Z_g - Z_*\right) $$

$$Z^\prime = Z* - Z_g \implies \frac{dZ^\prime}{dt} = \frac{dZ_*}{dt}$$

$$\frac{dZ^\prime}{dt} = -\frac{1}{M_*}\frac{dM_*}{dt} Z^\prime $$

Assume sSFR = const and $Z_g$ = const over integration time

$$\frac{dZ^\prime}{Z^\prime} = -{\rm sSFR} dt $$

$$\ln\left(\frac{Z^\prime(t)}{Z^\prime_0}\right) = -({\rm sSFR}) t $$

$$Z^\prime(t) = Z^\prime_0 e^{-{\rm sSFR} t} $$

```python
class metals():
    
    def __init__(self, Zstar0, Zgaas0, sSFR0, dt, deltaZ = 0.02):
        '''Specify deltaZ if you want to change 'miss' mode delta
        
        '''
        # Initial values
        self.Zstar   = Zstar0
        self.Zgas    = Zgas0
        self.sSFR    = sSFR0
        self.t       = 0
        self.T       = 0
        self.targetZ = Zgas0
        self.deltaZ  = deltaZ
    
        # Constants
        self.dt = dt
        
        # Helpers
        self.nSteps = 1e8
        
        self.Zstars  = np.zeros(int(self.nSteps))#[self.Zstar]
        self.Zgasses = np.zeros(int(self.nSteps))#[self.Zgas]
        self.ts      = np.zeros(int(self.nSteps))#[self.t]
        
        self.Zstars [0] = self.Zstar
        self.Zgasses[0] = self.Zgas
        self.ts     [0] = self.t
        
        # Continuous Zgas changing option
        self.mode = '' # No default set
    
    def randomModeParams(self,mu=0.0,sigma=1e-2):
        self.mu = mu
        self.sigma = sigma
    
    def updateTarget(self,newTarget):
        self.targetZ = newTarget
        if (self.mode == "HIT"):
            ### TODO: FIX THIS ###
            if ((self.T - self.t)/self.dt) <= 1:
                    self.deltaZ = (self.targetZ - self.Zgas)
            else:
                self.deltaZ = (self.targetZ - self.Zgas) / ((self.T - self.t)/self.dt)
        elif (self.mode == "INSTANT"):
            self.Zgas = newTarget
        if (self.mode == "RANDOM"):
            self.randomModeParams() # Initialize, can be changed later
        
    def changeZgas( self ):
        funcFromMode = {
            "HIT":self.hitMode,
            "MISS":self.missMode,
            "RANDOM":self.randomMode,
            "INSTANT":self.instantMode
        }
        return funcFromMode[self.mode](self.Zgas)
    
    def hitMode(self, Zgas):
        # Updates Zgas by linear fit, will exactly hit new target
        return Zgas + self.deltaZ 
        
    def missMode(self, Zgas):
        # Updates Zgas by set deltaZ, not guarenteed to hit new target
        newZgas = Zgas
        if (Zgas < self.targetZ):
            # Avoid oscilatory solutions
            if (Zgas + self.deltaZ > self.targetZ):
                newZgas = self.targetZ
            else:
                newZgas = Zgas + self.deltaZ
        elif (Zgas > self.targetZ):
            if (Zgas - self.deltaZ < self.targetZ):
                newZgas = self.targetZ
            else:
                newZgas = Zgas - self.deltaZ
        return newZgas
    
    def randomMode(self,Zgas):
        # Randomly draw a deltaZ
        deltaZ = self.sigma*np.random.randn() + self.mu
        
        return Zgas + deltaZ
    
    def instantMode(self,Zgas):
        # Instant change in Zgas
        return Zgas
    
    def getZprime( self, ZS, ZG ):
        return ZS - ZG
    
    def step( self, t, sSFR, dt ):
        newT     = t + dt
        newZstar = self.Zstar + sSFR*(self.Zgas - self.Zstar)*dt

        return newT, newZstar
        
    def doIntegration( self ):
        while (self.t < self.T):            
            if (self.index == self.nSteps):
                print('Index larger than nSteps!!!')
                break
            newT, newZstar = self.step( self.t, self.sSFR, self.dt )
            
            newZgas = self.changeZgas()
            
            self.Zstars [self.index] = newZstar
            self.Zgasses[self.index] = newZgas
            self.ts     [self.index] = newT
            
            self.Zstar = newZstar
            self.Zgas  = newZgas
            self.t     = newT
            
            self.index += 1
        
    def run( self, ZgasArr, TArr, mode="" ):
        
        MODES = ['HIT','MISS','RANDOM','INSTANT']
        
        if (mode.upper() not in MODES):
            raise TypeError('Invalid Mode. ' + f"Modes are: {MODES}")
        else:
            self.mode = mode.upper()
        print(self.mode)
        self.index = 0
        for index, newZgas in enumerate(ZgasArr):
            if (index % 1000) == 0:
                print(index)
            self.T += TArr[index]
            nStepsToAdd = int((self.T - self.t)/self.dt) + 1
#             self.Zstars  = np.concatenate( (self.Zstars ,np.zeros( nStepsToAdd )) )
#             self.Zgasses = np.concatenate( (self.Zgasses,np.zeros( nStepsToAdd )) )
#             self.ts      = np.concatenate( (self.ts     ,np.zeros( nStepsToAdd )) )
            
            self.updateTarget(newZgas)
            self.doIntegration()
        
        # Remove last step (is an extra 0)
        self.Zstars  = self.Zstars [self.ts > 0]
        self.Zgasses = self.Zgasses[self.ts > 0]
        self.ts      = self.ts     [self.ts > 0]
        
        self.Zprime = self.getZprime( self.Zstars, self.Zgasses )
        print( f"Integration complete in {len(self.ts)} steps" )
```

```python
HubbleTimez0  = 13.6e9 # z=0
HubbleTimez2  =  3.1e9 # z=2
HubbleTimez5  =  1.1e9 # z=5
HubbleTimez10 = 0.48e9 # z=10

HUBBLE_TIMES = [HubbleTimez2,HubbleTimez5,HubbleTimez10]
indexToZ     = {0:"2",1:"5",2:"10"}

GAS_MZR_VAL  = 0.0
STAR_MZR_VAL = 0.0

Zstar0 = STAR_MZR_VAL
Zgas0  = GAS_MZR_VAL

dt     = 1e6

num_Zgas_jumps = 10000
Zgs = 0.3 * np.random.randn( num_Zgas_jumps ) + GAS_MZR_VAL
# Too wide
Ts  = [np.random.exponential( scale=0.1*Hubble_Time, size=num_Zgas_jumps ) for Hubble_Time in HUBBLE_TIMES]

deltaZ = 1e-3
```
